```python
# X = diabetes_df.drop(labels='Progression', axis=1) 
# y = diabetes_df['Progression'] OR:
X = diabetes.data
y = diabetes.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

<h1 style="color: green">KNeighborsRegressor<h1>

```python
from sklearn.neighbors import KNeighborsRegressor

kn_reg = KNeighborsRegressor(n_neighbors=1)
kn_reg.fit(X_train, y_train)

kn_reg5 = KNeighborsRegressor(n_neighbors=5)
kn_reg5.fit(X_train, y_train)

kn_reg10 = KNeighborsRegressor(n_neighbors=10)
kn_reg10.fit(X_train, y_train)
```

```python
y_pred = kn_reg.predict(X_test)
y_pred5 = kn_reg5.predict(X_test)
y_pred10 = kn_reg10.predict(X_test)
print('predictions on test set with 1 neighbor:\n{}'.format(y_pred))
print('-------------------------------------------')
print('predictions on test set with 5 neighbors:\n{}'.format(y_pred5))
print('-------------------------------------------')
print('predictions on test set with 10 neighbors:\n{}'.format(y_pred10))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse5 = mean_squared_error(y_test, y_pred5)
r2_5 = r2_score(y_test, y_pred5)

mse10 = mean_squared_error(y_test, y_pred10)
r2_10 = r2_score(y_test, y_pred10)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
print('-------------------------------------------')
print(f"Mean Squared Error: {mse5}")
print(f"R^2 Score: {r2_5}")
print('-------------------------------------------')
print(f"Mean Squared Error: {mse10}")
print(f"R^2 Score: {r2_10}")
```

```python
print('accuracy on the trained dataset: {:.2f}'.format(kn_reg.score(X_train, y_train)))
print('accuracy on the tested dataset: {:.2f}'.format(kn_reg.score(X_test, y_test)))
print('-------------------------------------------')
print('accuracy on the trained dataset: {:.2f}'.format(kn_reg5.score(X_train, y_train)))
print('accuracy on the tested dataset: {:.2f}'.format(kn_reg5.score(X_test, y_test)))
print('-------------------------------------------')
print('accuracy on the trained dataset: {:.2f}'.format(kn_reg10.score(X_train, y_train)))
print('accuracy on the tested dataset: {:.2f}'.format(kn_reg10.score(X_test, y_test)))
```

```python
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 5, 10], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    predictions = reg.predict(X_test)
    # scatter
    ax.plot(y_test, predictions, 'o')
    # perfect fit
    ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)

    ax.set_title(
        "{} neighbor(s)\n test R^2: {:.2f}".format(n_neighbors, reg.score(X_test, y_test))
    )
    ax.set_xlabel('actual target / true target')
    ax.set_ylabel('predicted target / model predictions')

axes[0].legend(['predicted vs. actual', 'perfect fit'], loc='best')
plt.show()
```
