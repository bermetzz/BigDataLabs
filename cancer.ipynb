<h1 style="color: green">Explore the data<h1>

```python
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
```

<h1 style="color: green">Dividing the data into training and test sets<h1>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(breast_cancer['data'], breast_cancer['target'], random_state=0)
```

```python
print('X_train array shape: {}'.format(X_train.shape))
print('y_train array shape: {}'.format(y_train.shape))
print('................................................')
print('X_test array shape: {}'.format(X_test.shape))
print('y_test array shape: {}'.format(y_test.shape))
```

```python
import pandas as pd

# creating df
df = pd.DataFrame(breast_cancer['data'], columns=breast_cancer['feature_names'])
df.head(5)
```

<h1 style="color: green">Exploring data using visualization. Selecting the necessary params
graphs<h1>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# preprocessing step(to normalize the feature values)
scaler = StandardScaler()
scaler.fit(df)

scaled_data = scaler.transform(df)

pca = PCA(n_components=2)
pca.fit(scaled_data)

x_pca = pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=breast_cancer['target'])
plt.xlabel('first principle component')
plt.ylabel('second principle component')
```

```python
import numpy as np
import seaborn as sns

# creating df and concat 'data' and 'target' together
cancer_df = pd.DataFrame(np.c_[breast_cancer['data'], breast_cancer['target']],
                         columns=np.append(breast_cancer['feature_names'], ['target']))
```

```python
# pair plot of some features of df
sns.pairplot(cancer_df, hue='target',
             vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
# showing the total count of malignant and benign tumor patients 
sns.countplot(data=cancer_df, x='target')
plt.title('malignant and benign tumor patients count')
```

```python
# find a correlation between each feature and target
plt.figure(figsize=(10, 10))
sns.heatmap(cancer_df.corr(), annot=True, cmap='coolwarm', linewidths=2)
```

```python
# taking the correlation of each feature with the target and visualizing bar plot

# second df with dropped target
cancer_df2 = cancer_df.drop(['target'], axis=1)

# visualize correlation bar plot
plt.figure(figsize=(14, 5))
ax = sns.barplot(x=cancer_df2.columns, y=cancer_df2.corrwith(cancer_df['target']))
ax.tick_params(labelrotation=90)
```

<h1 style="color: green">Building a model using the k-nearest means method. Testing on
arbitrary dataset<h1>

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
import numpy as np

X_new = np.array([[46.03, 72.12, 0.06, 13.53, 12.26, 0.04, 12.33, 9.75, 75.11, 34.22, 0.01, 98.21, 11.36, 46.03, 72.12,
                   0.06, 13.53, 12.26, 0.04, 12.33, 9.75, 75.11, 34.22, 0.01, 98.21, 11.36, 0.00, 0.12, 0.11, 9.00]])
print('X_new array shape: {}'.format(X_new.shape))
```

<h1 style="color: green">Model prediction<h1>

```python
prediction = knn.predict(X_new)
print('prediction: {}'.format(prediction))
print('predicted label: {}'.format(breast_cancer['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print('prediction for test set: \n{}'.format(y_pred))
```

<h1 style="color: green">Accuracy (model quality assessment)<h1>

```python
print('accuracy of test set: {:.2f}'.format(np.mean(y_pred == y_test)))
print('accuracy of test set: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# confusion matrix
confusion = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(confusion)

vis = sns.heatmap(confusion, annot=True, cmap='nipy_spectral')
vis
```
